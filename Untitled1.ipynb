```python
import os
import json
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'/home/ning/personachat'

file1 = 'test_none_original.txt'
        
def data_preprocess(file_path,file_name):
    with open(os.path.join(file_path,file_name),'r') as f:
        context = f.read() 
    lines = context.split('\n')
    conv = [lines[i].split('\t')[0:2] for i in range(len(lines))]
    
    con = []
    start = 0
    end = 0
    for i,c in enumerate(conv):
        if c[0]=='':
            continue
        if c[0][0] == '1':
            start = i
        if conv[i] > conv[i+1]:
            end = i+1
            con.append(conv[start:end])
    for c in con:
        for t in c:
            t[0]=t[0][2:]
    
    conversations = []
    for c in con:
        conseq = []
        for t in c:
            conseq += t
        conversations.append(conseq)

    return conversations


def prepossing_data(conversations):
    df = pd.DataFrame(columns = ['utterance','causal_utterance ','emotion'])
    
    for conv in conversations:
        example_num = len(df.index)
        for i,utterance in enumerate(conv):
            causal_utterance = conv[i-1]
            emotion = None
            
            if i:
                df.loc[i+example_num] = [utterance,causal_utterance,emotion]
            else:
                df.loc[i+example_num] = [utterance,'None',emotion]
            
    return df

ccc = data_preprocess(file_path,file1)
df = prepossing_data(ccc)


with open(os.path.join(file_path,file1.split('_')[0]+'_df.pkl'),'wb') as f:
    pickle.dump(df,f)
```
